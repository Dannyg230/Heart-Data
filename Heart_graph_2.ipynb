```python
#read the data
heart = pd.read_csv(r'C:\Users\danny\Downloads\heart.csv')
saturation = pd.read_csv(r'C:\Users\danny\Downloads\o2Saturation.csv')
heart.head()
heart.info()
heart.describe()
```

```python
#number of heart attacks related to age
plt.figure(figsize=(14, 10))
sns.countplot(x = 'age',data = heart, palette = 'coolwarm')
```

```python
#chnace of heart of attack related to blood pressure and cholestrol
plt.figure(figsize = (12,10))
sns.scatterplot(x = 'chol', y = 'thalachh', data = heart, hue = "output", s = 70, palette =['green', 'red'], legend = 'full')
plt.legend(["Less Chance of Heart Attack", "Risk of Heart Attack"],fontsize = 12)
plt.xlabel('Cholestrol')
plt.ylabel('Resting blood Pressure')
```
